```python
tests = []
with open("test17.txt") as f:
    for lines in f:
        for coord in lines.split(" "):
            if coord != "" :
                if '\n' in coord:
                    tests.append(coord.replace("\n",""))
                else :
                    tests.append(coord)

for i,coords in enumerate(tests):
    x,y = coords.split(",")
    tests[i] = (int(x), int(y))

tests
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_traj(list):
    target_area = [(139, -148), (139, -89), (187, -148), (187, -89)]
#     target_area = [(20, -10), (20, -5), (30, -10), (30, -5)]
    target_x, target_y = zip(*target_area)
    traj_x, traj_y = zip(*list)
    plt.scatter(target_x, target_y)
    plt.scatter(0,0, c='r')
    plt.scatter(traj_x, traj_y, c='g')
    plt.ylim([-150, 10])
    plt.show()


def probe_route(vx, vy, xlim, ylim):
    xmin, xmax = xlim
    ymin, ymax = ylim
    Vx, Vy = vx, vy
    res = [(0,0)]
    x,y = res[0]
    while x <= xmax and y >= ymin:
        x += Vx
        y += Vy
        if Vx > 0 :
            Vx -= 1
        elif Vx < 0:
            Vx += 1
        Vy -= 1
        res.append((x,y))
        if xmin <= x <= xmax and ymin <= y <= ymax:
#             print("Hit!")
            return res, True
#     print("Miss")
    return res, False
    
def count_route(xlim, ylim):
    xmin, xmax = xlim
    ymin, ymax = ylim
    x,y = 0, ymin
    vx, vy = 0, 0
    res = []
    while x < xmax+1:
        if probe_route(x, y, xlim, ylim)[1]:
            res.append((x,y))
        if y > np.abs(ymin)-1:
            y = ymin
            x += 1
        else :
            y += 1
    return res

def compare_route(l1, l2):
    res = []
    t = 0
    f = 0
    for x in l1:
        if not x in l2:
            res.append(x)
        else :
            t += 1
    for x in l2:
        if x not in l1:
            f += 1 
    print(t, " true")
    print(f, " false")
    return res
```
