```python
#11.1
import numpy as np

INPUT = np.zeros([10,10])
with open("input11.txt") as f:
    X = f.read().replace('\n',' ').strip(' ').split(' ')
    for i in range(len(X)):
        for digit in range(len(X[i])):
            INPUT[i][digit] = int(X[i][digit])

def add1xy(tab, x, y):
    try :
        if tab[y][x] == 0:
            pass
        else :
            tab[y][x] += 1
    except :
        pass

def add1(tab):
    for y in range(len(tab)):
        for x in range(len(tab[y])):
            tab[y][x] += 1

def addadj(tab, x, y):
    if x > 0 :
        if y > 0 :
            add1xy(tab,  x-1, y-1)
        add1xy(tab, x-1, y+1)
        add1xy(tab, x-1, y)
    if y > 0 :
        add1xy(tab, x,  y-1)
        add1xy(tab, x+1,y-1)
    add1xy(tab, x, y+1)
    add1xy(tab, x+1, y)
    add1xy(tab, x+1, y+1)


def step(tab, x, y, clk):
    if tab[y][x] >= 10 :
        tab[y][x] = 0
        addadj(tab, x, y)
        clk = True
    return clk

def stepall(tab, clk):
    clk = False
    for y in range(len(tab)):
        for x in range(len(tab[y])):
            clk = step(tab, x, y, clk)
    return clk

def countf(tab):
    flash = 0
    for lines in tab:
        flash += np.count_nonzero(lines == 0)
    return flash

INPUT
#9818 sad
#1413 ff
#1653 low ff
#1656 sad
```
