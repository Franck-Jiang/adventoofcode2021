```python

def bin (param) :
    res = 0
    for i in range (0, len(param)) :
        res += 2**i * int(param[len(param) - i - 1])   
    return res


def crit(liste, index) :
    if len(liste) == 1 :
        return liste, liste
    one = 0
    oneL =[]
    zero = 0
    zeroL = []

    for i in range(len(liste)) :
        if liste[i][index] == '1' :
            one += 1
            oneL.append(liste[i])
        else : 
            zero += 1
            zeroL.append(liste[i])
    if one >= zero :
        return oneL, zeroL
    else :
        return zeroL, oneL 

res = []
with open("input3.txt") as f:
    for lines in f :
        res.append(lines)

oxy = res
co2 = res
for i in range(12) :
    oxy = crit( oxy, i)[0]
    co2 = crit( co2, i)[1]

print(bin( str( int(oxy[0]) ) ) * bin( str( int(co2[0]) )) )
```
