```python
import copy

def pad(input):
    res = copy.copy(input)
    while len(input)%4 != 0:
        input += '0'
    return input

def read_literal_value(input):
    res = ""
    data = input
    fb = data[0]
#     print(fb)
    data = data[1:]
    
    while True :
        res += data[:4]
#         print("A", data[:4])
        data =  data[4:]
    
        if fb == '0':
            return b2d(res), data
        fb = data[0]
#         print("B", data[0])
        data = data[1:]
    return 
            

def read_transmission(bin_input): 
    #read the first 3 bits of a binary input
    stored_data = copy.copy(bin_input)
    
    version = int(stored_data[:3],2)
    stored_data = stored_data[3:]
    print("version =", version)
    
    typeID = b2d(stored_data[:3])
    stored_data = stored_data[3:]
    print("type id =", typeID)
    
    if typeID == 4:
        data = pad(stored_data)
        val, stored_data = read_literal_value(data)
        print("literal value =", val)
        
    else:
        lengthID = stored_data[0]
        stored_data = stored_data[1:]
        print("lengthID = ", lengthID)
        
        if lengthID == '0':
            nb_bit = b2d(stored_data[:15])
            print("nb_bit =", nb_bit)
            
            stored_data = stored_data[15:]
            subpack = stored_data[:nb_bit]
            stored_data = stored_data[nb_bit:]
            while len(subpack) > 6 and ('1' in subpack) :                
                print("--subpack")
                try :
                    vers, subpack = read_transmission(subpack)
                except :
                    print( subpack )
                    raise Exception
                version += vers
                print("subpack--")

        elif lengthID == '1':
            nb_sub = b2d(stored_data[:11])
            stored_data = stored_data[11:]
            for i in range(nb_sub):
                print("--subpack")
                vers, stored_data = read_transmission(stored_data)
                version += vers
                print("subpack--")
            
    return version, stored_data


```

```python
def calc_list(l : list):
    
    if len(l) == 2 :
        if type(l[1]) == int :
            return l[1]
        else :
            return calc_list(l[1])
    if l[0] == '+' :
        res_sum = 0
        for n in l[1:] :
            if type(n) == int:
                res_sum += n
            else :
                res_sum += calc_list(n)
        return res_sum
    
    if l[0] == '*' :
        res_prod = 1
        for n in l[1:] :
            if type(n) == int:
                res_prod *= n
            else :
                res_prod *= calc_list(n)
        return res_prod
    
    if l[0] == 'min' :
        res_min = l[1]
        for n in l[2:]:
            if n < res_min :
                res_min = n
        return res_min
    
    if l[0] == 'max' :
        res_max = l[1]
        for n in l[2:]:
            if n > res_max :
                res_max = n
        return res_max
    
    if l[0] == '>' :
        if type(l[1]) == int :
            n = l[1]
        else :
            n = calc_list(l[1])
        if type(l[2]) == int :
            m = l[2]
        else:
            m = calc_list(l[2])
        return 1 if n > m else 0
    
    if l[0] == '<' :
        if type(l[1]) == int :
            n = l[1]
        else :
            n = calc_list(l[1])
        if type(l[2]) == int :
            m = l[2]
        else:
            m = calc_list(l[2])
        return 1 if n < m else 0
    
    if l[0] == '=' :
        if type(l[1]) == int :
            n = l[1]
        else :
            n = calc_list(l[1])
        if type(l[2]) == int :
            m = l[2]
        else:
            m = calc_list(l[2])
        return 1 if n == m else 0
    raise Exception("idk")
            
```
